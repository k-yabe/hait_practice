```python
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
import numpy as np
np.random.seed(0)
x_1=np.random.normal(50,10,100)
y_1 = x_1 + np.random.normal(0,5,100)
```

```python
x_2 = np.random.normal(30,5,100)
y_2 = 2.5*x_2 + np.random.normal(0,10,len(x_2))
```

```python
plt.scatter(x_1,y_1,color='red')
plt.scatter(x_2,y_2,color='blue')
plt.title('The relaationship between x and y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.scatter(x_1,y_1,color='red',marker='s')
plt.scatter(x_2,y_2,color='blue',marker='*')
plt.title('The relationship between x and y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.scatter(x_1,y_1,color='red',marker='s',label='data1')
plt.scatter(x_2,y_2,color='blue',marker='x',label='data2')
plt.title('The relationship between x and y')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='lower right')
plt.show()
```

```python
plt.scatter(x_1,y_1,color='red',marker='s',label='date1')
plt.scatter(x_2,y_2,color='blue',marker='x',label='data2')
plt.title('The relationship between x and y')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper right')
plt.xticks([20,50,80])
plt.yticks([20,50,80,110])
plt.show()
```

```python
plt.scatter(x_1,y_1,color='red',marker='s',label='data1')
plt.scatter(x_2,y_2,color='blue',marker='x',label='data2')
plt.title('The relationship between x_1 and y_1')
plt.xlabel('feature')
plt.legend(loc='upper right')
plt.xticks(np.arange(10,80,5))
plt.yticks(np.arange(10,130,10))
plt.show()
```

```python
plt.scatter(x_1,y_1,color='red',marker='s',label='data1')
plt.scatter(x_2,y_2,color='blue',marker='x',label='data2')
plt.title('The relationship between x and y')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper right')
plt.xticks(np.arange(10,80,5))
plt.yticks(np.arange(10,130,10))
plt.grid()
plt.show()
```

```python
x_3 = np.arange(-10,10,0.1)
y_3 = 0.001 * (x_3**3 + x_3**2 + x_3 + 1) 
```

```python
plt.plot(x_3,y_3)
plt.show()
```

```python
plt.plot(x_3, y_3, label='cubic')      # 凡例のためにlabelで凡例名をつける
plt.title('a cubic function')                # グラフのタイトルを追加
plt.xlabel('x')                                       # x軸のラベルを表示
plt.ylabel('y')                                       # y軸のラベルを表示
plt.legend(loc='upper left')               # 凡例を表示
plt.xticks(np.arange(-10, 11, 1))     # x軸のラベルを指定
plt.yticks(np.arange(-1, 1.5, 0.25)) # y軸のラベルを指定
plt.grid()                                               # グリッド線の表示
plt.show()                                             # 指示した内容でグラフを描画
```

```python
plt.plot(x_3, y_3, label='cubic')      # 凡例のためにlabelで凡例名をつける
plt.title('a cubic function')                # グラフのタイトルを追加
plt.xlabel('x')                                       # x軸のラベルを表示
plt.ylabel('y')                                       # y軸のラベルを表示
plt.legend(loc='upper left')               # 凡例を表示
plt.xticks(np.arange(-10, 11, 1))     # x軸のラベルを指定
plt.yticks(np.arange(-1, 1.5, 0.25)) # y軸のラベルを指定
plt.grid()                                        # グリッド線の表示
plt.hlines([0],-10,10,linestyles='dashed', color='gray')
plt.vlines([0],-1,1,linestyles='dashed',color='gray')
plt.show()                                             # 指示した内容でグラフを描画
```
